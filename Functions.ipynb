## Y = 2X-5

```python
# Return evenly spaced values within a given interval
x = np.arange(-4,6,0.1899)
y = 2*(x)-5
# lets add some noise
y_noise = np.random.normal(size = x.size)
# Draw random samples from a normal (Gaussian) distribution size is same as our 
ydata = y+y_noise
plt.plot(x,y,"b")
plt.plot(x,ydata,'bo',color = 'green')
plt.xlabel("x")
plt.ylabel("y")
```

## Lets draw the Gaussian distribution

```python
import matplotlib.pyplot as plt
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

Non-linear regression is a method to model the non-linear relationship between the independent variables $x$ and the dependent variable **y** Essentially any relationship that is not linear can be termed as non-linear, and is usually represented by the polynomial of M degrees. 

$$\ y = a x^3 + b x^2 + c x + d \$$

we will plot logarihmic function 

$$\ log (a x^3 + b x^2 + c x + d) \$$


```python
x = np.arange(-6,4,0.1)
y = 2*(x**3)+3*(x**2)+2*x-7
y_noise = 30*np.random.normal(size = x.size)
y_data = y+y_noise
plt.plot(x,y,'r')
plt.plot(x,y_data,'bo',color ='green',alpha = 1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```



# $$Y = x^2$$

```python
x = np.arange(-6,6,0.1)
y = x**2
y_noise = 2*np.random.normal(size = x.size)
y_data = y+y_noise
plt.plot(x,y,'r')
plt.plot(x,y_data,'bo',color ='green',alpha = 1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# **$$ Y = a + b c^X$$**

```python
x = np.arange(-6,4,0.1)
y = 3+2*4**x
y_noise = 30*np.random.normal(size = x.size)
y_data = y+y_noise
plt.plot(x,y,'r')
plt.plot(x,y_data,'bo',color ='green',alpha = 1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# $$ y = log(x)$$

```python
x = np.arange(-6,6,0.1)
y = np.log(x)
y_noise = np.random.normal(size = x.size)
y_data = y+y_noise
plt.plot(x,y,'r')
plt.plot(x,y_data,'bo',color ='green',alpha = 1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Sigmoidal/Logistic

$$ Y = a + \frac{b}{1+ c^{(X-d)}}$$

```python
x = np.arange(-6,6,0.1)
y = 2 - 3/(1+np.power(3,x-3))
y_noise = 0.5*np.random.normal(size = x.size)
y_data = y+y_noise
plt.plot(x,y,'r')
plt.plot(x,y_data,'bo',color ='green',alpha = 1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
